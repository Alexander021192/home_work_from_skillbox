```python
# Рисуем графики сразу же
%matplotlib inline
# подключаем все необходимые в будущем библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from datetime import datetime
import dateutil.parser
import numpy as np

#plt.style.use('ggplot')  # Красивые графики
plt.rcParams['figure.figsize'] = (15, 5)  # Размер картинок
```

```python
#загружаем все таблицы csv
students = pd.read_csv('Data/students.csv')
courses = pd.read_csv('Data/courses.csv')
course_contents = pd.read_csv('Data/course_contents.csv')
progresses = pd.read_csv('Data/progresses.csv')
progress_phases = pd.read_csv('Data/progress_phases.csv')

```

```python
courses.rename(columns={'id':'course_id'}, inplace = True) # переименовал колонку для соединения таблиц методом merge
progresses.rename(columns={'id':'progress_id'}, inplace = True)
merged_1 = courses.merge(course_contents) # соединил таблицу courses и course_contents
merged_2 = progresses.merge(progress_phases) # соединил таблицу progresses и progress_phases
main_df = merged_1.merge(merged_2) # и соединил все таблицы вместе

```

```python
main_df.rename(columns={'title':'course_title', 'field':'course_field' }, inplace = True) # чуть переименовал столбцы, для понимания
main_df.drop(columns='Unnamed: 0', inplace = True)
main_df.to_csv('main_df.csv') # сохранил основной датасет. можно подгружать его в дальнейших заданиях.
main_df[main_df.student_id == '768c2987a744c51ce64a5993a2a94eaf'] # проверка основного датасета
```

```python
course_group = list(main_df.groupby('course_title'))
counts_courses = len(main_df.groupby('course_title'))
#sum_count_courses = len(main_df.groupby('course_title')) #общее количество курсов в датасете
print(f"общее количество курсов в датасете: {counts_courses}")
# создадим также список имен курсов. Пригодится для будущих циклов.
courses = [course_group[i][0] for i in range(counts_courses)]
print(f'список всех курсов: {courses}')
```

```python
# посчитаем количество модулей на каждом курсе
# отберем из основного датасета нужные нам столбцы
module_numbers_for_course = main_df[['course_title','module_title']]
# убираем дубликаты, чтобы при подсчете модулей не мешали.
module_numbers_for_course = module_numbers_for_course.drop_duplicates() 
# группируем по курсам, и считаем количество модулей
module_numbers_for_course = module_numbers_for_course.groupby('course_title').count() 
module_numbers_for_course.rename(columns={'module_title':'count_module'}, inplace=True)

module_numbers_for_course #количество модулей на каждом курсе
```

```python
# посчитаем количество уроков в каждом модуле на каждом курсе,
# отберем из основного датасета нужные нам столбцы
count_lesson_for_module = main_df[['course_title','module_title','lesson_number']]
# убираем дубликаты, чтобы при подсчете уроков не мешали.
count_lesson_for_module = count_lesson_for_module.drop_duplicates()
# группируем по курсам и модулям,  считаем количество уроков
count_lesson_for_module = count_lesson_for_module.groupby(['course_title','module_title']).count()
count_lesson_for_module.rename(columns={'lesson_number':'count_lessons'})
```

```python
# посчитаем медианное количество уроков в модуле на каждом курсе
# отберем из основного датасета нужные нам столбцы
median_lesson_for_module = main_df[['course_title','module_title','lesson_number']]
# убираем дубликаты, чтобы при подсчете уроков не мешали.
median_lesson_for_module = median_lesson_for_module.drop_duplicates()
# считаем количество уроков в каждом модуле
median_lesson_for_module = median_lesson_for_module.groupby(['course_title','module_title']).count()
# и теперь считаем медианное количество уроков
median_lesson_for_module = median_lesson_for_module.groupby('course_title').median()
median_lesson_for_module.rename(columns={'lesson_number':'median_count_lessons'})
```

```python
# посчитаем количество учеников на каждом курсе
# отберем из основного датасета нужные нам столбцы
count_students_for_course = main_df[['course_title','student_id']]
# убираем дубликаты, чтобы при подсчете студентов не мешали.
count_students_for_course = count_students_for_course.drop_duplicates()
# группируем по курсам и  считаем количество студентов
count_students_for_course = count_students_for_course.groupby('course_title').count()
count_students_for_course.rename(columns={'student_id':'count_student'}, inplace=True)

count_students_for_course #количество студентов на каждом курсе
```

```python
#посчитаем минимальный, максимальный, средний, медианный возраст студентов

# текущая дата
year_now = datetime.now().year
# удалим из датасета студентов без даты рождения
age_students = students.dropna()
# отберем из датасета нужные нам столбцы
age_students = age_students[['id','birthday']]
# переименуем столбец для большего понимания значения
age_students.rename(columns={'id':'student_id'}, inplace=True)
# преобразуем строковое представления даты в тип(класс) Datetime. для дальнейших расчётов
age_students['birthday'] = age_students['birthday'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d").date()) 
# создаем новую колонку, в которой считаем возрас студентов
age_students['age'] = age_students['birthday'].apply(lambda x: year_now - x.year)
# удалим колонку с датой рождения. Он больше не нужен
age_students.drop('birthday', axis='columns', inplace=True)

# как и в любых данных, в наших есть опечатки с датой рождения.
#поэтому уберем вручную точно некорректные данные. лиц младше 14 лет. и старше 100
age_students = age_students.query("age<100 & age > 14")

# считаем все необходимые нам характеристики. 
age_students.describe()
```

```python
#посчитаем минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
# отберем из датасета нужные нам столбцы
age_students_for_course = main_df[['course_title','student_id']]
# убираем дубликаты
age_students_for_course = age_students_for_course.drop_duplicates()
# соединим уже готовый датасет с возрастом каждого студента
age_students_for_course = age_students_for_course.merge(age_students)
# далее просто группируем по курсу и считаем нужные нам характеристики
age_students_for_course.groupby('course_title')['age'].agg(['min','max','mean','median'])
```

(m) Постройте bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы

```python
df = count_students_for_course.reset_index().sort_values(by='count_student') #количество студентов на каждом курсе
#Построим bar-chart, отражающий количество студентов на каждом курсе
ax = df.plot.bar(x='course_title', y='count_student', rot=90)
ax.set_xlabel('курсы')
ax.set_ylabel('количество студентов')
#легенда мне здесь не нужна
ax.legend([])
#убрал серый фон
ax.patch.set_visible(False)
```

(m) Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов на каждом курсе. График должен иметь заголовок. Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере, к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐


```python
# определяем область рисования графика и его размер.
fig, ax = plt.subplots(figsize=(13, 7))

# список цветов для графиков(для каждого курса)
colors = ['black', 'silver', 'firebrick', 'sandybrown', 'gold', 'olivedrab',
          'chartreuse', 'green', 'cyan', 'darkblue', 'coral', 'orange',
          'yellow', 'red', 'greenyellow']

# рисую столбец для каждого курса, чтобы у всех был отдельный цвет на графике и легенде. 
for i in range(len(colors)):
    #  df - таблица количество студентов на каждом курсе       
    # рисую столбец для каждого курса, чтобы у всех был отдельный цвет на графике и легенде. 
    ax.barh(i, df['count_student'].iloc[i], label = df['course_title'].iloc[i], color=colors[i])
    
    # указываем характеристики легенды: местоположение, прозрачность.
    ax.legend(loc = 'lower right', ncol=1, fontsize=13, framealpha=0.1)
    # подписываем график
    ax.set_title('Количество студентов на каждом курсе')
    
    #подписываем оси x and y 
    ax.set_xlabel('Количество студентов')
    ax.set_ylabel('Курсы')
    # убираем подписи для оси Y
    ax.yaxis.set_major_formatter (ticker.NullFormatter())
    #указываем характеристики сетки графика:цвет, толщина
    ax.grid(which='major',color = 'k', linewidth = 0.4, linestyle = '-')
    #убрал серый фон
    ax.patch.set_visible(False)
    
# рисуем медиану количества студентов.
ax.vlines(df['count_student'].median(), -1, 15, color='purple')

plt.tight_layout()
```

Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом по дате начала первой домашней работы.



```python
# выберем те строки, где есть домашнее задание
increace_students = main_df.query('is_homework == True')
# отбираем необходимы столбцы для расчета из основного датасета
increace_students = increace_students[['course_title','module_number','start_date']]

#НАМ ЖЕ НУЖНЫ ТОЛЬКО ДОМАШКИ ПЕРВОГО МОДУЛЯ
increace_students = increace_students.query("module_number==1")
# создал новую колонку время, созданную из start_date. Переведел в удобный вид: год/месяц
increace_students['start_date'] = increace_students['start_date'].apply(lambda x: x[0:7])


increace_students = increace_students.rename(columns={'start_date':'date'}).sort_values(by=['course_title','date'])
increace_students
```

```python
# группируем таблицу по курсам и по дате(год/месяц) и производим подсчет
increace_students_counts = increace_students.groupby(['course_title','date']).count()
# сортируем по курсам и дате
increace_students_counts = increace_students_counts.sort_values(by=['course_title','date'], ascending=True).reset_index()
increace_students_counts = increace_students_counts.rename(columns={'module_number':'increase'})
# хочу вставить строки с месяцами, где не было прироста. То есть при подсчете в некоторых месяцах должен стоять 0.
# создам таблицу со всеми месяцами, которые есть в курсе. 
all_month = main_df[['course_title','start_date']].rename(columns={'start_date':'date'})
all_month['date'] = all_month['date'].apply(lambda x: x[0:7])
all_month = all_month.drop_duplicates().sort_values(by=['course_title','date'])
# соединим обе таблицы. Чтобы в пропущенных месяцах стоял Nan, который затем поменяю на 0.
increace_students_counts = all_month.merge(increace_students_counts, how='left')
increace_students_counts = increace_students_counts.fillna(0)
increace_students_counts.head()
```

Учитывать, что выполнение домашнего задания может перетекать из одного месяца в другой (такие дз надо включать в общее число прогрессов для всех месяцев, которые покрывает срок выполнения этих дз)

```python
# отберем необходимые столбцы из основного датасета
progress_homework = main_df[['course_title','is_homework','start_date','finish_date']]
# выберем все значения, где есть домашняя работа
progress_homework = progress_homework.query("is_homework==True")
# перенес столбец is_homework в конец таблицы
progress_homework = progress_homework[['course_title','start_date','finish_date','is_homework']]
progress_homework
```

```python
'''Учитываем работы, которые находятся в статусе,
отличном от 'done'. Для удобства их можно учесть только в одном первом месяце.
Таким образом я поставил, будто они были сделаны, в том же месяце, что и начаты.'''
progress_homework['finish_date'] = np.where(pd.isna(progress_homework['finish_date']),progress_homework['start_date'],progress_homework['finish_date'])
progress_homework
```

```python
# переводим строковые значения в столбце start_date в тип datetime, для расчетов
progress_homework['start_date'] = progress_homework['start_date'].apply(lambda x: x[0:7])
progress_homework['finish_date'] = progress_homework['finish_date'].apply(lambda x: x[0:7])
# на всякий случай сделаем копию. датасета
progress_homework_1 = progress_homework.copy()
```

```python
progress_homework_1.head()
```

```python
# сгруппируем и посчитаем количество прогрессов 
# без учёта перетекания пока что
progress_homework_1 = progress_homework_1.groupby(['course_title','start_date','finish_date'])['is_homework'].count()
progress_homework_1 = pd.DataFrame(progress_homework_1).reset_index()
# переименнуем столбцы
progress_homework_1 = progress_homework_1.rename(columns={'is_homework':'count_progress', 'start_date':'time'})
```

```python
progress_homework_1.loc[progress_homework_1['course_title'] == 'Excel Базовый' ].head(6) 
```

```python
# создаем функцию для учета перетекания домашнего задания из месяца в месяц.
def my_df_for_count_progress(df):
    for i, elem_finish in enumerate(df['finish_date']):
        if elem_finish>df['time'].iloc[i]:
            start_date = df['time'].iloc[i]
            for j, elem_start in enumerate(df['time'].iloc[i:]):
                if elem_finish >= elem_start > start_date:
                    df['count_progress'].iloc[j+i] += df['count_progress'].iloc[i]
                    start_date=elem_start
                elif elem_finish < elem_start:
                    break
    df = pd.DataFrame(df.groupby(['course_title','time'])['count_progress'].sum()).reset_index()
    return df
```

```python
# посчитаем число прогрессов с учётом перетекания выполнения дом.задания из месяца в месяц
# создам пустой датафрейм, куда буду сохранять операции просчета прогресса с учтом перетекания по каждому курсу
progress_homework_sum = pd.DataFrame()
for i, course in enumerate(courses):
    df = progress_homework_1.query(f"course_title == '{course}'")
    # счиатем прогрессы с учетом перетекания на отдельном курсе
    df = my_df_for_count_progress(df)
    # и складываем потом все вместе
    progress_homework_sum = pd.concat([progress_homework_sum, df])
progress_homework_sum
```

```python
# создадим функцию, которая будет считать время прохождения модуля
def my_timedelta(x):
    result = x['finish_date'].iloc[-1] - x['start_date'].iloc[0]
    return result
# сгруппируем значения и  применим функция для расчета времени прохождения модуля    
module_transit_time_1 = module_transit_time.groupby(['course_title','student_id','module_title','module_number']).apply(lambda x: my_timedelta(x))
```

```python
# подготовим данные для расчета конверсии.
# отберем нужные столбцы для рассчетов. будем считать количество студентов сдавших дз
conversion_1 = main_df[['course_title','module_number','lesson_number','finish_date']].dropna()
# сначала посчитал для всех дз.
finish_count = conversion_1.groupby(['course_title','module_number','lesson_number']).count()
# выбрал только количество по последнему дз в модуле
finish_count = finish_count.groupby(['course_title','module_number']).apply(lambda x: x['finish_date'].iloc[-1])
```

```python
# подготовим данные для расчета конверсии.
# отберем нужные столбцы для рассчетов. будем считать количество студентов приступивших к дз
conversion_2 = main_df[['course_title','module_number','lesson_number','start_date']].dropna()
# сначала посчитал для всех дз.
start_count = conversion_2.groupby(['course_title','module_number','lesson_number']).count()
# выбрал только количество по первому дз в модуле
start_count = start_count.groupby(['course_title','module_number']).apply(lambda x: x['start_date'].iloc[0])
```

```python
#у нас есть два Series, где посчитано количество студентов приступивших и сдавших дз, по первому и последнему дз соответственно
# соединим оба Series в один датасет conversion, для дальнеших расчетов
start_count = start_count.reset_index(name='start_count')
conversion = pd.DataFrame(start_count)
conversion['finish_count'] = list(finish_count)
conversion
```

```python
# создадим колонку, где будем записывать само значение конверсии
conversion['conversion'] = 0
# в цикле посчитаем конверсию.
for i in range(1, len(conversion)):
    # каждый раз когда меняется курс, конверсия для 1-го модуля должна оставаться нулевой. 
    if conversion['course_title'].iloc[i] == conversion['course_title'].iloc[i-1]:
        conversion['conversion'].iloc[i] = conversion['start_count'].iloc[i]/conversion['finish_count'].iloc[i-1]
    
conversion.head(12)    
```
